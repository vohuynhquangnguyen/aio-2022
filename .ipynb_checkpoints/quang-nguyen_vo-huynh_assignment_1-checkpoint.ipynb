1. **Viết 4 functions thực hiện tính thể tích của các khối sau:**
    * Hình lập phương $V = s^3$ ($s$ là độ dài của cạnh)
        * input function: $s$
        * output function: $V$
    * Hình hộp chữ nhật $V = l \times w \times h$ ($l$ là chiều dài, $w$ là chiều rộng, $h$ là chiều cao)
        * input function: $l, w, h$
        * output function: $V$
     * Hình trụ tròn: $V = \pi \times r^2 \times h$ ($r$ là bán kính đáy, $h$ là chiều cao)
        * input function: $r, h$
        * output function: $V$
      * Hình cầu: $V = \frac{4}{3}\times \pi \times r^3$ ($r$ là bán kính hình cầu)
        * input function: $r$
        * output function: $V$

NOTE: Tùy thuộc vào giá trị $\pi$ các bạn định nghĩa mà kết quả có thể sai lệch. Ví dụ tự định nghĩa $\pi$ = 3.14 hoặc dùng math.pi $\pi$ = 3.141592653589793

```python
##
# EXERCISE 1:
#
PI = math.pi

def calculate_cubical_volume(s, *args, **kwargs):
    volume = s ** 3
    return volume

def calculate_rectangular_solid_volume(l, w, h, *args, **kwargs):
    volume = l * w * h
    return volume

def calculate_cylindrical_solid_volume(r, h, *args, **kwargs):
    volume = PI * (r**2) * h
    return volume

def calculate_spherical_solid_volume(r, *args, **kwargs):
    volume = (4/3) * PI * (r**3)
    return volume
```

```python
##
# TEST:
#
volume = calculate_cubical_volume(s = 3)
print(volume)

volume = calculate_rectangular_solid_volume(l = 1, w = 2, h = 3)
print(volume)

volume = calculate_cylindrical_solid_volume(r = 1.2, h = 4)
print(volume)

volume = calculate_spherical_solid_volume(r = 1.3)
print(volume)

del volume
```

2. **Viết function thực hiện đánh giá classification model bằng F1-Score.**
  $$\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}$$
  $$\text{Recall} =\frac{\text{TP}}{\text{TP} + \text{FN}}$$
  $$\text{F1-score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$$
    * Input: function nhận 3 giá trị $\text{tp}$, $\text{fp}$, $\text{fn}$
    * Output: print ra kết quả của $\text{Precision}$, $\text{Recall}$, và $\text{F1-score}$.

NOTE: Đề bài yêu cầu các điều kiện sau
* Phải kiểm tra giá trị nhận vào `tp`, `fp`, `fn` là type int, nếu là type khác thì print ví dụ check fn là float, print `fn must be int!` và thoát hàm hoặc dừng chương trình.
* Yêu cầu `tp`, `fp`, `fn` phải đều lớn hơn 0, nếu không thì print `tp and fp and fn must be greater than zero!` và thoát hàm hoặc dừng chương trình
    * Input: function nhận 3 giá trị $\text{tp}$, $\text{fp}$, $\text{fn}$
    * Output: print ra kết quả của $\text{Precision}$, $\text{Recall}$, và $\text{F1-score}$.

NOTE: Đề bài yêu cầu các điều kiện sau
* Phải kiểm tra giá trị nhận vào `tp`, `fp`, `fn` là type int, nếu là type khác thì print ví dụ check fn là float, print `fn must be int!` và thoát hàm hoặc dừng chương trình.
* Yêu cầu `tp`, `fp`, `fn` phải đều lớn hơn 0, nếu không thì print `tp and fp and fn must be greater than zero!` và thoát hàm hoặc dừng chương trình

```python
##
# EXERCISE 2 (using 'assert' instead of 'if'):
#
def calculate_f1_score(tp, fp, fn, *args, **kwargs):
    assert type(tp) == int, 'tp must be int!'
    assert type(fp) == int, 'fp must be int!'
    assert type(fn) == int, 'fn must be int!'
    assert (tp > 0 ) and (fp > 0) and (fn > 0), 'tp and fp and fn must be greater than zero!'
    
    precision = tp / (tp + fp)
    print(f'precision is {precision}.')
    
    recall = tp / (tp + fn)
    print(f'recall is {recall}.')
    
    f1_score = 2 * (precision * recall) / (precision * recall)
    print(f'F1 score is {f1_score}.\n')
    
    return None
```

3. **Viết function mô phỏng theo 3 activation functions sau:**
    * Binary step function 
    \begin{equation*}
    f(x) = \begin{cases}
    0 \quad &\text{if} \, x < 0 \\
    1 \quad &\text{if} \, x \geq 0\\
    \end{cases}
    \end{equation*}

    * Sigmoid function
    \begin{equation*}
    f(x) = \frac{1}{1 + e^{-x}}
    \end{equation*}
    
    * Elu function
    \begin{equation*}
    f(x) = \begin{cases}
    \alpha(e^x - 1) \quad &\text{if} \, x <0\\
    x \quad &\text{if} \, x \geq 0\\
    \end{cases}
    \end{equation*}
    
    * Input
        * Người dùng nhập giá trị $x$
        * Người dùng nhập tên activation function chỉ có 3 loại (binary, sigmoid, elu)
    * Output: Kết quả $f(x)$ ($x$ khi đi qua activation fucntion tương ứng theo activation function name). Ví dụ nhập `x = 3`, `activation_function = 'binary'`, output: print `binary: f(3) = 1`
   
NOTE: Lưu ý các điều kiện sau:
* Dùng function `is_number` được cung cấp sẵn để kiểm tra x có hợp lệ hay không (vd: `x = 10, is_number(x)` sẽ trả về True ngược lại là False). Nếu không hợp lệ print `x must be a number` và dừng chương trình.
* Kiểm tra activation function name có hợp lệ hay không nằm trong 3 tên (binary, sigmoid, elu). Nếu không print `ten_function_user is not supported` (vd người dùng nhập `relu` thì print `relu is not supported`)
* Convert `x` sang float type
* Thực hiện theo công thức với activation name tương ứng. Print ra kết quả
* Dùng `math.e` để lấy số e
* $\alpha = 0.01$

```python
##
# EXERCISE 2 (using 'assert' instead of 'if'):
#
def is_number(n, *args, **kwargs):
    try:
        float(n)
    except ValueError:
        return False
    return True

def map_activation_function(x, function_name, *args, **kwargs):    
    assert is_number(x) == True, 'x must be a number!'
    assert ((function_name == 'binary') or (function_name == 'sigmoid') or (function_name == 'elu')), \
    f'{function_name} is not supported!'
    exp = math.e
    alpha = 0.01
    x = float(x)
    
    ##
    # Binary :
    #
    if (function_name == 'binary'):
        if (x >= 0):
            y = 1
        else:
            y = 0
        
    ##
    # Sigmoid:
    #
    elif (function_name == 'sigmoid'):
        y = 1 / (1 + exp ** x)
    
    ##
    # Elu:
    #
    elif (function_name == 'elu'):
        if (x >= 0):
            y = x
        else:
            y = alpha * (exp**(x) - 1)
            
    print(f'{function_name}: f({x}) = {y}')
```

```python
##
# TEST:
#
x = input('Input x = ')
function_name = input('Input activation function (binary|sigmoid|elu): ')
map_activation_function(x, function_name)
```

```python
##
# EXERCISE 2 (using only 'if'):
#
def is_number(n, *args, **kwargs):
    try:
        float(n)
    except ValueError:
        return False
    return True

def map_activation_function(x, function_name, *args, **kwargs):    
    if (is_number(x) != True):
        print('x must be a number!')
        return None
    
    if ((function_name != 'binary') and (function_name != 'sigmoid') and (function_name != 'elu')):
        print(f'{function_name} is not supported!')
        return None
    
    while(is_number(x) == True):
        exp = math.e
        alpha = 0.01
        x = float(x)

        ##
        # Binary :
        #
        if (function_name == 'binary'):
            if (x >= 0):
                y = 1
            else:
                y = 0

        ##
        # Sigmoid:
        #
        elif (function_name == 'sigmoid'):
            y = 1 / (1 + exp ** x)

        ##
        # Elu:
        #
        elif (function_name == 'elu'):
            if (x >= 0):
                y = x
            else:
                y = alpha * (exp**(x) - 1)

        print(f'{function_name}: f({x}) = {y}')
        break
```

4. **Viết function lựa chọn regression loss function để tính loss:**
    * $\text{MAE} = \frac{1}{n}\sum^n_{i = 1}|y_i − \hat{y}_i|$
    * $\text{MSE} = \frac{1}{n}\sum^n_{i = 1}(y_i − \hat{y}_i)^2$
    * $\text{RMSE} = \sqrt{\text{MSE}}$
    * $n$ chính là số lượng samples (`num_samples`), với $i$ là mỗi sample cụ thể. Ở đây các bạn có thể hiểu là cứ mỗi $i$ thì sẽ có 1 cặp $y_i$ là target và $\hat{y}_i$ là predict.
    * Input:
        * Người dùng nhập số lượng sample (`num_samples`) được tạo ra (chỉ nhận integer numbers)
        * Người dùng nhập loss name (`MAE, MSE, RMSE`). Chỉ cần MAE và MSE, bạn nào muốn làm thêm RMSE đều được.
        * Output: Print ra `loss name`, `sample`, `predict`, `target`, `loss`
            * loss name: là loss mà người dùng chọn
            * sample: là thứ tự sample được tạo ra (ví dụ `num_samples = 5`, thì sẽ có 5 samples và mỗi sample là `sample = 0`, `sample = 1`, `sample = 2`, `sample = 3`, `sample = 4`)
            * predict: là số mà model dự đoán (chỉ cần dùng random tạo random một số trong range [0,10])
            * target: là số target mà momg muốn mode dự đoán đúng (chỉ cần dùng random tạo random một số trong range [0,10])
            * loss: là kết quả khi đưa predict và target vào hàm loss
            * note: ví dụ `num_sample = 5` thì sẽ có 5 cặp predict và target.
            
Note: Các bạn lưu ý
* Dùng `.isnumeric()` method để kiểm tra num_samples có hợp lệ hay không (vd: `x = 10`, `num_samples.isnumeric()` sẽ trả về True ngược lại là False). Nếu không hợp lệ `print
'number of samples must be an integer number'` và dừng chương trình.
* Dùng vòng lặp for, lặp lại `num_samples` lần. Mỗi lần dùng module `random` tạo một con số ngẫu nhiên trong range [0.0, 10.0] cho predict và target. Sau đó đưa predict và target vào loss function và print ra kết quả mỗi lần lặp.
* Dùng `random.uniform(0,10)` để tạo ra một số ngẫu nhiên trong range [0,10].
* Giả xử người dùng luôn nhập đúng loss name MSE, MAE, và RMSE (đơn giản bước này để các bạn không cần check tên hợp lệ).
* Dùng `abs()` để tính trị tuyệt đối ví dụ `abs(-3)` sẽ trả về 3.
* Dùng `math.sqrt()` để tính căn bậc 2.


```python
##
# EXERCISE 4:
#
def compute_loss(num_samples, loss_name, *args, **kwargs):
    assert (num_samples.isnumeric() == True), 'Number of samples must be an integer number!'
    num_samples = int(num_samples)
    Y_pred = []
    Y_target = []
    LOSS = []
    
    for i in range(num_samples):
        y_pred = random.uniform(0,10)
        Y_pred.append(y_pred)
        
        y_target = random.uniform(0,10)
        Y_target.append(y_target)
        
        ##
        # MAE pairing loss:
        #
        if (loss_name == 'MAE'):
            loss = abs(y_target - y_pred)
            LOSS.append(loss)    
        
        ## 
        # MSE pairing loss:
        #
        elif (loss_name == 'MSE'):
            loss = (y_target - y_pred) ** 2
            LOSS.append(loss)
        
        ## 
        # RMSE pairing loss:
        #
        elif (loss_name == 'RMSE'):
            loss = (y_target - y_pred) ** 2
            LOSS.append(loss)
            
        print(f'loss name: {loss_name}; sample: {i}; pred: {y_pred}; target: {y_target}; loss: {loss}')
    
    ##
    # Final loss:
    #
    if (loss_name != 'RMSE'):
        final = sum(LOSS) / len(LOSS)
    else:
        final = math.sqrt(sum(LOSS) / len(LOSS))
    print(f'final {loss_name}: {final}')
    
    return None
            
```

```python
num_samples = input('Input number of samples (integer number) which are generated: ')
loss_name = input('Input loss name: ')
compute_loss(num_samples, loss_name)
```

5. **(OPTIONAL) Viết các function thực hiện tam giác Pascal và dãy số Fibonacci:**

**Pascal’s Triangle**
* Input: level là số lượng hàng
* Output: Print ra kết quả từng hàng theo Pascal’s Triangle

**Fibonacci Sequence**
* Input: length là độ dài của chuỗi Fibonacci
* Output: Print ra kết quả các elemet trong chuỗi Fibonacci
* Các bạn nên tìm hiểu các công thức có thể thực hiện được Pascal’s Triangle và Fibonacci
Sequence

```python
##
# EXERCISE 5: 
# 
def compute_pascal_triangle(n_rows, *args, **kwargs):
    assert (n_rows > 0), 'Number of rows must be a positive integer'
    
    for row in range(n_rows):
        for i in range(row+1):
            numerator = math.factorial(row)
            denominator = math.factorial(i) * math.factorial(row - i)
            combination = int(numerator / denominator)
            print(f'{combination}', end = ' ')
        print('\n')
    return None

def compute_Fibonacci(length, *args, **kwargs):
    assert (length > 0), 'Length must be a positve integer!'
    
    Fibonacci_list = []
    if (length == 1):
        n = 0
        Fibonacci_list.append(n)
        print(Fibonacci_list)
        return None
    else:
        n_0 = 0
        n_1 = 1
        Fibonacci_list.append(n_0)
        Fibonacci_list.append(n_1)
        
        count = 0
        while (count < length):
            n_th = n_0 + n_1
            n_0 = n_1
            n_1 = n_th
            Fibonacci_list.append(n_th)
            count += 1
        print(Fibonacci_list)
        return None
```

```python
##
# TEST: 
#
compute_pascal_triangle(5)
compute_Fibonacci(5)
```

6. **(OPTIONAL) Viết 4 functions để ước lượng các hàm số sau:**
    * Input: $x$ (số muốn tính toán) và $n$ (số lượng bậc muốn xấp xỉ)
    * Output: Kết quả ước lượng hàm tương ứng với x. Ví dụ hàm `cos(x = 0)` thì `output = 1`
    * $x$ là radian
    * n là số nguyên dương $n \geq 0$
    * Các bạn nên viết một hàm tính giai thừa riêng

```python
def compute_factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

def approximate_sin(x, n, *args, **kwargs):
    assert (n >= 0), 'x must be a positive integer'
    
    sin_list = []
    for i in range(n):
        numerator = x ** (2 * i + 1)
        denominator = compute_factorial(2 * i + 1)
        sin = (-1) ** i * (numerator / denominator)
        sin_list.append(sin)
    
    final = sum(sin_list)
    print(final)
    
def approximate_cos(x, n, *args, **kwargs):
    assert (n >= 0), 'x must be a positive integer'
    
    cos_list = []
    for i in range(n):
        numerator = x ** (2 * i)
        denominator = compute_factorial(2 * i)
        cos = (-1) ** i * (numerator / denominator)
        cos_list.append(cos)
    
    final = sum(cos_list)
    print(final)
    
def approximate_sinh(x, n, *args, **kwargs):
    assert (n >= 0), 'x must be a positive integer'
    
    sinh_list = []
    for i in range(n):
        numerator = x ** (2 * i + 1)
        denominator = compute_factorial(2 * i + 1)
        sinh = numerator / denominator
        sinh_list.append(sinh)
    
    final = sum(sinh_list)
    print(final)

def approximate_cosh(x, n, *args, **kwargs):
    assert (n >= 0), 'x must be a positive integer'
    
    cosh_list = []
    for i in range(n):
        numerator = x ** (2 * i)
        denominator = compute_factorial(2 * i)
        cosh = numerator / denominator
        cosh_list.append(cosh)
    
    final = sum(cosh_list)
    print(final)
        
```

```python
approximate_sin(3.14, 10)
approximate_cos(3.14, 10)
approximate_sinh(3.14, 10)
approximate_cosh(3.14, 10)
```

7. **(OPTIONAL) Cho một số nguyên dương n, viết phương trình đảo ngược thứ tự các vị trí trong số n. Chỉ dùng while (hay for) và những phép toán cơ bản như $+, -, \times, /, \%, ...$ :**
    * Input: n là một dãy số nguyên dương.
    * Output: Đảo ngược vị trí các số trong n. Ví dụ `input: 12345678910`, `output: 1987654321`

NOTE: Các bạn chú ý các điều kiện sau
* Không được ép kiểu sang string
* Chỉ sử dụng while hoặc for loop

```python
##
# EXERCISE 7:
#
def reverse_number(number, *args, **kwargs):
    reverse = 0
    while(number > 0):
        last_digit = number % 10
        reverse = last_digit + (reverse * 10)
        number = number // 10
    print(reverse)
```
