1. **Problem 1 (Calculus):** Given
$$
f(x) = 4x^3 − 48x^2 + 180x − 5
$$
Find its stationary points, then indicate whether they are maximum, or minimum and find these maximum and
minimum values?

***

**Answer:**
1. First derivative of $f(x)$:
$$
f'(x) = 12x^2 - 96x + 180
$$
2. Find points where $f'(x) = 0$:
$$
f'(x) = 0 \\
\iff 12x^2 - 96x + 180 = 0 \\
\iff x = 3 \quad \lor \quad x = 5 \\
$$
3. Second derivative of $f(x)$:
$$
f'(x) = 24(x - 4)
$$
4. Substitute $x = 3$ and $x =5$ to $f'(x)$:
\begin{equation*}
\begin{cases}
x = 3 \implies 24(3 - 4) = -24 < 0 \\
x = 5 \implies 24(5 - 4) = 24 > 0 \\
\end{cases}
\end{equation*}

5. Based on the results, we can conclude that:
    * $x = 3$ is the maximum.
    * $x = 5$ is the minimum.

2. **Problem 2 (Calculus):**
    * Find $\frac{d}{dx}g(f(x))$? Given $f(x) = 5 − 6x$, $g(u) = u^5$
    * Determine area between two functions: the upper function $y = 25 + 5x$, and the lower function $y = 19 + 6x + x^2$. Two curves will intersect at $x = −3$ and $x = 2$.

***

**Answer:**
* For the first question:
    * The derivative of $f(x)$:
    $$
    f'(x) = \frac{df}{dx} = -6
    $$
    * The derivative of $g(u)$:
    $$
    g'(u) = \frac{dg}{du} = 5u^4
    $$
    * Apply chain rule, the derivative of $\frac{d}{dx}g(f(x))$ is:
    $$
    \frac{dg}{dx} = \frac{dg}{du} \frac{du}{df} \frac{df}{dx} = 5\times(5 − 6x)^4\times(-6) = -30(5 − 6x)^4
    $$
    where $\frac{du}{df} = 1 \iff u = f = 5 - 6x$.

* For the second question:
    * Find the integral of $y = 25 + 5x$ in the range [-3, 2]:
    $$
    \int^2_{-3}(25 + 5x) = \Big (25\times(2) + \frac{(5\times(2)^2)}{2} \Big )- \Big (25\times(-3) + \frac{(5\times(-3)^2)}{2} \Big ) = 112.5
    $$
    * Find the integral of $y = 19 + 6x + x^2$ in the range [-3,2]:
    $$
    \int^2_{-3}19 + 6x + x^2 = \Big (19 \times (2) + 3 \times (2)^2 + \frac{(2)^3}{3} \Big ) - \Big (19 \times (-3) + 3 \times (-3)^2 + \frac{(-3)^3}{3} \Big )\\
     = \frac{275}{3}
    $$
    * The area between to functions is:
    $$
    A = \int^2_{-3}(25 + 5x) - \int^2_{-3}19 + 6x + x^2 = 20.8\bar{3} = 20\frac{5}{6}
    $$

7. **Problem 7 (Probability):** A box contains 10 black balls and 2 red balls.
* Two balls are selected at random without replacement (select a ball then it is not given back to the box). Find the probability that the two balls are of
different colors.
* Three balls are selected at random with replacement. Find the probability that all three balls are black.

*** 

**Answer:**
* For the first question.
    * Since we do not care about the order, i.e., red + black = black + red, the probability of selecting two balls that are of difference colors is:
    $$
    P = \frac{C^1_{10} \times C^1_2}{C^2_{12}} = \frac{10\times 2}{66} = \frac{10}{33}
    $$


* For the second question.
    * The probability of selecting two balls that are of difference colors is:
    $$
    P = \frac{C^3_{10}}{C^3_{12}} = \frac{120}{220} = \frac{6}{11}
    $$

8. **Problem 8 (Probability):** A COVID-19 diagnostic test is reported that the test result is positive, given that the
person has COVID-19, is 99%. And the test result is positive, given that the person does not have COVID-19, is only
2% . It is known that 60% of all people have COVID-19. If a patient test is positive, what is the probability that
they actually have the COVID-19.

***

**Answer**:
* We have:
$$
\begin{cases}
P(\text{positive}|\text{have}) = 0.99 \\
P(\text{positive}|\bar{\text{have}}) = 0.02 \\
P(\text{have}) = 0.6 \implies P(\bar{\text{have}}) = 1.0 - 0.6 = 0.4 \\
\end{cases}
$$
* Thus:
$$
P(\text{have}|\text{positive}) = \frac{P(\text{have})P(\text{positive}|\text{have})}{P(\text{positive})}
$$
* Meanwhile:
$$
P(\text{positive}|\text{have}) = \frac{P(\text{positive} \land \text{have})}{P(\text{have})} \\ 
\iff P(\text{positive} \land \text{have}) = 0.99\times0.6 = 0.594
$$

$$
P(\text{positive}|\bar{\text{have}}) = \frac{P(\text{positive} \land \bar{\text{have}})}{P(\bar{\text{have}})} \\ 
\iff P(\text{positive} \land \bar{\text{have}}) = 0.02\times0.4 = 0.008
$$
* Moreover,
$$
P(\text{positive} \land \text{have}) + P(\text{positive} \land \bar{\text{have}}) = P(\text{positive}) = 0.594 + 0.008 = 0.602
$$
* Therefore,
$$
P(\text{have}|\text{positive}) = \frac{0.6\times 0.99}{0.602} \approx 0.9867
$$
