cái và `value` là số lần xuất hiện
* Input: một từ
* Output: dictionary đếm số lần các chữ xuất hiện
* Note: Giả sử các từ nhập vào đều có các chữ cái thuộc [a-z] hoặc [A-Z]
 
(b) Viết function đọc các câu trong một file txt, đếm số lượng các từ xuất hiện và trả về một dictionary với key `là` từ và `value` là số lần từ đó xuất hiện.
* Input: Đường dẫn đến file txt
* Output: dictionary đếm số lần các từ xuất hiện
* Note:
    * Giả sử các từ trong file txt đều có các chữ cái thuộc [a-z] hoặc [A-Z]

```python
##
# EXERCISE 1 (a):
#

def count_characters_from_word_v1(word, *args, **kwargs):
    alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

    word = word.lower()
    character_list = []
    occurrence_list = []

    ##
    # Count the number of occurrences of a given letter:
    #
    for _ , value in enumerate(alphabet):
        
        count = 0
        for _ , c in enumerate(word):
            if c == value:
                count += 1
        
        if (count > 0):
            occurrence_list.append(count)
            character_list.append(value)
        
        ##
        # Build the dictionary:
        #
        output_dictionary = dict(zip(character_list, occurrence_list))
            
    return output_dictionary

def count_characters_from_word_v2(word, *args, **kwargs):
    
    ##
    # Count the number of occurrences of a given letter:
    #
    counts = dict()
    for character in word:
        if  character in counts:
            counts[character] += 1
        else:
            counts[character] = 1

    return counts
```

2. Thực hiện xây dựng Stack bằng dictionary và list data type. Viết một function nhận 4 arguments: `method_name`, `stack`, `capacity`, `value` để mô phỏng lại cách hoạt động
của stack
    * Input: `method_name`, `stack`, `capacity`, `value`
    * Outcome: thực hiện theo `method_name` và ghi kết quả vào `stack` (input ở trên)
    * Các loại method cần thực hiện trong `method_name` argument:
        * `CreateStack` dùng để khởi tạo `stack`
        * `IsEmptyStack` kiểm tra `stack` có đang rỗng
        * `IsFullStack` kiểm tra `stack` đã full chưa
        * `PopStack` loại bỏ top element và đưa giá trị vào RetValue" trong `stack` dictionary
        * `PushStack` add thêm `value` (input argument) vào trong `stack`
        * `GetTopStack` lấy giá trị của top element trong stack đưa vào `TopValue` (không làm thay đổi stack)
    * `stack argument`: là một dictionary chứa các thông tin cần thiết để tạo thành một stack data. Sau đây là các thông tin cơ bản yêu cầu phải có, các bạn có thể thêm các thông tin tùy ý để
thực hiện được stack data cho riêng mình.
        * `Stack` (key) : [] (value là kiểu list chứa các element trong stack)
        * `Capacity` (key) : int (value là một số nguyên dương thể hiện size tối đa cho phép của
stack)
        * `IsEmptyStack` (key) : boolean (value là kiểu boolean `True` nếu stack rỗng và `False` nếu stack có chứa ít nhất một element)
        * `IsFullStack` (key) : boolean (value là kiểu boolean True nếu số lượng element trong stack bằng `Capacity`, `False` nếu số lượng element trong stack nhỏ hơn Capacity
        * `RetValue` (key) : cùng type với element type (value là gíá trị trả về khi thực hiện `PopStack` method)
        * `TopValue` (key) : cùng type với top element type (value là giá trị trả về của top element khi thực hiện `GetTopStack` method)
    * `capacity argument`: size tối đa của stack và chỉ được truyền vào khi đi cùng `CreateStack`
method
    * `value argument`: là gía trị sẽ được push vào stack và chỉ được truyền vào khi đi cùng `PushStack` method

```python
##
# EXERCISE 2:
#
def CreateStack(stack, capacity, *args, **kwargs):
    assert(type(stack) is dict), "createStack method expects `stack` argument!"
    assert(capacity is not None), "createStack method expects `capacity` argument!"

    stack['Stack'] = []
    stack['Capacity'] = capacity
    stack['TopIdx'] = -1
    stack['IsEmptyStack'] = None
    stack['IsFullStack'] = None
    stack['RetValue'] = None
    stack['TopValue'] = None

    return stack

def IsEmptyStack(stack, *args, **kwargs):
    assert(type(stack) is dict), "IsEmptyStack method expects `stack` argument!"

    if (stack['TopIdx'] == -1):
            stack['IsEmptyStack'] = True
    else:
            stack['IsEmptyStack'] = False    
    return stack

def IsFullStack(stack, *args, **kwargs):
    assert(type(stack) is dict), "IsFullStack method expects `stack` argument!"

    if stack['TopIdx'] == stack['Capacity'] - 1:
        stack['IsFullStack'] = True
    else:
        stack['IsFullStack'] = False

    return stack

def PopStack(stack, *args, **kwargs):
    assert(type(stack) is dict), "PopStack method expects `stack` argument!"
    IsEmptyStack(stack)
    assert(stack['IsEmptyStack'] != True), "PopStack method expects non-empty `stack`!"
    stack['RetValue'] = stack['Stack'].pop()
    stack['TopIdx'] -= 1
    return stack

def PushStack(stack, value, *args, **kwargs):
    assert(type(stack) is dict), "PushStack method expects `stack` argument!"
    assert(value != None), "PushStack method expects valid `value` argument!"
    assert(stack['IsFullStack'] != True), "PopStack method expects non-full `stack`!"
    stack['Stack'].append(value)
    stack['TopIdx'] += 1

    return stack

def GetTopStack(stack, *args, **kwargs):
    assert(type(stack) is dict), "GetTopStack method expects `stack` argument!"
    IsEmptyStack(stack)
    assert(stack['IsEmptyStack'] != True), "PopStack method expects non-empty `stack`!"
    stack['TopValue'] = stack['Stack'][stack['TopIdx']]
    return stack

def simStack(method_name, stack = None, capacity = None, value = None, *args, **kwargs):
    if method_name == 'CreateStack':
        result = CreateStack(stack, capacity)

    elif method_name == 'IsEmptyStack':
        result = IsEmptyStack(stack)

    elif method_name == 'IsFullStack':
        result = IsFullStack(stack)

    elif method_name == 'PopStack':
        result = PopStack(stack)

    elif method_name == 'PushStack':
        result = PushStack(stack, value)
    
    elif method_name == 'GetTopStack':
        result = GetTopStack(stack)
    
    return result
```

```python
##
# TEST:
#
example = \
    {
        'Stack': [],
        'Capacity' : None,
        'IsEmptyStack' : None,
        'IsFullStack' : None,
        'RetValue' : None,
        'TopValue' : None,
    }

result = simStack(stack = example, method_name = 'CreateStack', capacity = 5)
print(result)

result = simStack(stack = result, method_name = 'PushStack', value = 1)
result = simStack(stack = result, method_name = 'PushStack', value = 2)
print(result['Stack'])

result = simStack(stack = result, method_name = 'IsFullStack')
print(result['IsFullStack'])

result = simStack(stack = result, method_name = 'GetTopStack')
print(result['TopValue'])

result = simStack ( stack = result , method_name = 'PopStack')
print(result['RetValue'])

result = simStack(stack = result, method_name = 'GetTopStack')
print(result['TopValue'])

result = simStack(stack = result, method_name = 'PopStack')
print(result['RetValue'])

result = simStack(stack = result, method_name = 'IsEmptyStack')
print(result['IsEmptyStack'])
```

3. Thực hiện xây dựng Queue bằng dictionary và list data type. Viết một function nhận 4 arguments: `method_name`, `queue`, `capacity`, và `value` để mô phỏng lại cách hoạt
động của queue
* Input: `method_name`, `queue`, `capacity`, `value`
* Outcome: thực hiện theo method_name và ghi kết quả vào queue (input ở trên)
* Các loại method cần thực hiện trong `method_name` argument:
    * `CreateQueue` dùng để khởi tạo queue
    * `IsEmptyQueue` kiểm tra queue có đang rỗng
    * `IsFullQueue` kiểm tra queue đã full chưa
    * `Dequeue` loại bỏ first element và đưa giá trị vào `RetValue` trong queue dictionary
    * `Enqueue` add thêm value (input argument) vào trong queue
    * `GetFirstValue` lấy giá trị của first element trong queue đưa vào `FirstValue` (không làm thây đổi queue)
* queue argument: là một dictionary chứa các thông tin cần thiết để tạo thành một queue data. Sau đây là các thông tin cơ bản yêu cầu phải có, các bạn có thể thêm các thông tin tùy ý để thực hiện được queue data cho riêng mình.
    * `Queue` (key) : [] (value là kiểu list chứa các element trong queue)
    * `Capacity` (key) : int (value là một số nguyên dương thể hiện size tối đa cho phép của queue)
    * `IsEmptyQueue` (key) : boolean (value là kiểu boolean True nếu queue rỗng và False nếu queue có chứa ít nhất một element)
    * `IsFullQueue` (key) : boolean (value là kiểu boolean True nếu số lượng element trong
queue bằng Capacity, False nếu số lượng element trong queue nhỏ hơn Capacity
    * `RetValue` (key) : cùng type với element type (value là gíá trị trả về khi thực hiện`Dequeue` method)
    * `FirstValue` (key) : cùng type với first element ty

```python
##
# EXERCISE 3:
#
def CreateQueue(queue, capacity, *args, **kwargs):
    assert(type(queue) is dict), "CreateQueue method expects `stack` argument!"
    assert(capacity is not None), "CreateQueue method expects `capacity` argument!"

    queue['Queue'] = []
    queue['Capacity'] = capacity
    queue['IsEmptyQueue'] = None
    queue['IsFullQueue'] = None
    queue['RetValue'] = None
    queue['FirstValue'] = None

    return queue

def IsEmptyQueue(queue, *args, **kwargs):
    assert(type(queue) is dict), "IsEmptyQueue method expects `queue` argument!"
    queue['IsEmptyQueue'] = (len(queue['Queue']) == 0)
    
    return queue

def IsFullQueue(queue, *args, **kwargs):
    assert(type(queue) is dict), "IsFullQueue method expects `queue` argument!"
    queue['IsFullQueue'] = (queue['Capacity'] == len(queue['Queue']))

    return queue

def Dequeue(queue, *args, **kwargs):
    assert(type(queue) is dict), "Dequeue method expects `queue` argument!"
    IsEmptyQueue(queue)
    assert(queue['IsEmptyQueue'] != True), "Dequeue method expects non-empty `queue`!"
    queue['RetValue'] = queue['Queue'].pop(0)
    return queue

def Enqueue(queue, value, *args, **kwargs):
    assert(type(queue) is dict), "Enqueue method expects `queue` argument!"
    assert(value != None), "Enqueue method expects valid `value` argument!"
    IsFullQueue(queue)
    assert(queue['IsFullQueue'] != True), "Enqueue method expects non-full `stack`!"
    queue['Queue'].append(value)

    return queue

def GetFirstValue(queue, *args, **kwargs):
    assert(type(queue) is dict), "GetFirstValue method expects `queue` argument!"
    IsEmptyQueue(queue)
    assert(queue['IsEmptyQueue'] != True), "GetFirstValue method expects non-empty `queue`!"
    queue['FirstValue'] = queue['Queue'][0]
    return queue

def simQueue(method_name, queue = None, capacity = None, value = None, *args, **kwargs):
    if method_name == 'CreateQueue':
        result = CreateQueue(queue, capacity)

    elif method_name == 'IsEmptyQueue':
        result = IsEmptyQueue(queue)

    elif method_name == 'IsFullQueue':
        result = IsFullQueue(queue)

    elif method_name == 'Dequeue':
        result = Dequeue(queue)

    elif method_name == 'Enqueue':
        result = Enqueue(queue, value)

    elif method_name == 'GetFirstValue':
        result = GetFirstValue(queue)

    return result
```

```python
##
# TEST:
#
example = \
    {
    'Queue' : [],
    'Capacity' : None,
    'IsEmptyQueue' : None,
    'IsFullQueue' : None,
    'RetValue' : None,
    'FirstValue' : None
}

result = simQueue(queue = example, method_name = 'CreateQueue', capacity = 5)
print(result)

result = simQueue(queue = result, method_name = 'Enqueue', value = 1)
result = simQueue(queue = result, method_name = 'Enqueue', value = 2)
print(result['Queue'])

result = simQueue(queue = result, method_name = 'IsFullQueue')
print(result['IsFullQueue'])

result = simQueue(queue = result, method_name = 'GetFirstValue')
print(result['FirstValue'])

result = simQueue(queue = result, method_name = 'Dequeue')
print(result['RetValue'])

result = simQueue(queue = result, method_name = 'GetFirstValue')
print(result['FirstValue'])

result = simQueue(queue = result, method_name = 'Dequeue')
print(result['RetValue'])

result = simQueue(queue = result, method_name = 'IsEmptyQueue')
print(result['IsEmptyQueue'])
```

4. **Stock Span Problem**: Cho trước một list giá của một loại cổ phiếu trong N ngày. Viết function
đi tìm stock span cho mỗi ngày trong list trên.
* Input: list chứa giá cổ phiểu của N ngày (một ngày là một giá)
* Output: trả về một list là stock span của từng ngày
* Khuyến khích các bạn sử dụng stack, và nếu có thể thì sử dụng stack của các
bạn đã thực hiện ở bài 2
* Stock Span: của một ngày hiện tại chính là số lượng ngày liên tiếp tối đa bắt đầu từ ngày
hiện tại và cho đến những ngày trước trong quá khứ có giá cổ phiếu <= giá cổ phiếu của
ngày hiện tại
* Ví dụ:
    *  Input: `[100, 60, 70, 65, 80, 85]`
    *  Output: `[1, 1, 2, 1, 4, 5]`


```python
##
# EXERCISE 4
#
def stock_span(rates):
    
    ##
    # 1. Create empty stack, and set default case:
    #
    stockspan = []
    stockspan.append(1)

    stack = [] 
    stack.append(0)

    for idx in range(1,len(rates)):
      ##
      # 2. Pop elements out of stack until either: 
      # 1) The stack gets empty, or 
      # 2) the rate at stack top turns out to be larger than the rate at the current element
      while rates[idx] > rates[stack[-1]]:
        stack.pop()

        if len(stack) == 0:
          break
      
      ##
      #  3. Set the stockspan values.
      #
      if len(stack) > 0:
          stockspan.append(idx - stack[-1])
      else:
          stockspan.append(idx + 1) 

      stack.append(idx)
    
    return stockspan
```

```python
##
# TEST:
#
rates = [100, 60, 70, 65, 80, 85]
stockspan = stock_span(rates)
print(stockspan)
```

5. Cho trước một range với N số nguyên dương (từ 1 đến N). Viết function biến đổi các
số thập phân trong range sang dạng binary string và trả về một dictionary với key
là số thập phân và value là binary dạng string
* Input `N` là số nguyên dương thể hiện range từ 1 đến N
* Output: một dictionary với key là số thập phân và value là binary dạng string tương ứng
* Khuyến khích các bạn sử dụng queue, và nếu có thể thì sử dụng queue của các
bạn đã thực hiện ở bài 3
* Các bạn có thể tìm hiểu thêm binary number system tại [link](https://www.mathsisfun.com/binary-number-system.html)

```python
##
# EXERCISE 5:
#
def generate_binary_string(N):
    decimal_list = [i for i in range(1, N + 1)]
    binary_list = [bin(i).replace('0b', '') for i in decimal_list]

    output_dictionary = dict(zip(decimal_list, binary_list))
    return output_dictionary
```
