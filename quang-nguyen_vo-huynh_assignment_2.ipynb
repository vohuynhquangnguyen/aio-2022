1. **Viết function nhận input vào là 1 chuỗi ký tự (string), biến đổi string theo yêu cầu
print ra màn hinh.**
    * Input: input_string là một chuỗi ký tự bất kỳ
    * Output: print ra theo điều kiện sau
        * Kiểm tra xem nếu chuỗi ký tự chỉ chứa số nguyên dương thì print "The string is number
only" và thoát hàm
        * Nếu không chỉ chứa số nguyên dương thì print ra string theo các yêu cầu sau:
        * Các chữ cái trong string đều là dạng viết thường
        * Các chữ cái trong string đều viết hoa
        * Chỉ chữ cái đầu tiên mỗi từ viết hoa

````python
##
# EXERCISE 1:
#

def transform_string(input_string, *args, **kwargs):
    ##
    # Check whether the string is digit-only:
    #
    count = 0
    for c in input_string:
        if c.isdigit() == True:
            count += 1
    if count == len(input_string):
        print('The string is number only!')
        return None
    else:
        string_type_1 = input_string.lower()
        print(string_type_1)

        string_type_2 = input_string.upper()
        print(string_type_2)

        string_type_3 = input_string.title()
        print(string_type_3)
        
        return None
````

2. **Cho một chuỗi coded_str chứa các ký tự đã được mã hóa theo cách sau `k[string]`,
trong đó `k` là số lần string trong dấu `[]` được lặp lại. Điều kiện: `coded_str` là các chuỗi
các ký tự tiếng Anh và string chỉ chứa các ký tự `[a-z]` và `[A-Z]` không có ký tự đăc biệt
và không phải là số. Nếu là số thì bắt buộc phải là `k` và đứng trước `[]`. Viết function
giải mã `coded_str` và trả về string đã được giải mã:**
* Note: `k` trong range `[1,9]`
* Note: Giả xử `coded_str` luôn nhâp đúng theo yêu cầu (Đơn giản hóa vấn đề để các bạn không
cần check nhiều trường hợp lỗi)
* Input: `coded_str` là string đã được mã hóa bởi `k` và `[]`
* Output: Trả về 1 string được giải mã bằng cách lập lại k lần các string trong [] và giữ nguyên
các ký tự còn lại
    * Input: `coded_str = "2[abc]3[cd]ef"`
    * Output: `"abcabccdcdcdef"`
    * Input: `coded_str = "abc3[cd]xyz"`
    * Output: `"abccdcdcdxyz"`

````python
##
# EXERCISE 2:
#

def decode_string(coded_string, *args, **kwargs):
    ##
    # Locate where left square brackets are:
    #
    left_brackets_positions = []    
    for pos, char in enumerate(coded_string):
        if(char == '['):
            left_brackets_positions.append(pos)

    ##
    # Locate where right square brackets are:
    #
    right_brackets_positions = []
    for pos, char in enumerate(coded_string):
        if(char == ']'):
            right_brackets_positions.append(pos)

    ##
    # Repeat the string within the square brackets
    #
    string_list = []
    for idx in range(len(left_brackets_positions)): # len(left_brackets_positions) = len(right_brackets_positions)
        left_bracket_pos = left_brackets_positions[idx]
        right_bracket_pos = right_brackets_positions[idx]

        string = coded_string[(left_bracket_pos + 1) : right_bracket_pos]
        string *= int(coded_string[left_brackets_positions[idx] - 1]) # Echo operation
        
        string_list.append(string)

    ##
    # Insert repeated strings into the original string:
    #
    result = coded_string
    for idx in range(len(string_list)): # len(string_list) = len(left_brackets_positions)
        left_bracket_pos = left_brackets_positions[idx]
        right_bracket_pos = right_brackets_positions[idx]
        string_to_be_replace = coded_string[(left_bracket_pos - 1) : (right_bracket_pos + 1)]
        result = result.replace(string_to_be_replace, string_list[idx])

    return result
````

````python
##
# TEST
#
coded_str = '2[abc]3[cd]ef'
result = decode_string(coded_str)
print(result)

coded_str = 'abc3[cd]xyz'
result = decode_string(coded_str)
print(result)
````

````python
##
# Example which shows content of 1 line in txt file (id, x_center, y_center, width, height):
#
example = '4 0.1104870557261957 0.6189416670904745 0.045474490406478114 0.1156995372018512'
example.split(' ')

##
# Example which shows how to extract id_class , x_center , y_center , width , height in 1 line of txt file:
#
id_class , x_center , y_center , width , height = example . split (' ')
print (f'id_class = {id_class}')
print (f'x_center = {x_center}')
print (f'y_center = {y_center}')
print (f'width = {width}')
print (f'height = {height}')

##
# Configure:
#
file1 = './dataset/contents/8.txt '
real_coor_file1 = './dataset/contents/8real.txt '
tl_br_coor_file1 = './dataset/contents/8tlbr.txt '
height1 = 1484
width1 = 2365

file2 = './dataset/contents/861.txt '
real_coor_file2 = './dataset/contents/861real.txt '
tl_br_coor_file2 = './dataset/contents/861tlbr.txt '
height2 = 541
width2 = 782
````

````python
##
# EXERCISE 3:
#
def show_content(file_path, *args, **kwargs):
    with open(file_path, 'r') as f:
        bounding_box_lines = f.readlines()

        for idx in range(len(bounding_box_lines)):
            line = bounding_box_lines[idx]
            print(line)

    return bounding_box_lines
    
def extract_content(file_path, *args, **kwargs):
    with open(file_path, 'r') as f:
        bounding_box_lines = f.readlines()

        id_class_list = []
        x_center_list = []
        y_center_list = []
        width_list = []
        height_list = []

        for idx in range(len(bounding_box_lines)):
            line = bounding_box_lines[idx]
            id_class, x_center, y_center, width, height = line.split(' ')
            id_class_list.append(id_class)
            x_center_list.append(x_center)
            y_center_list.append(y_center)
            width_list.append(width)
            height_list.append(height)

    return id_class_list, x_center_list, y_center_list, width_list, height_list

def compute_real_coord(id_class_list, x_center_list, y_center_list, width_list, height_list, width_img, height_img, *args, **kwargs):
    new_x_center_list = []
    new_y_center_list = []
    new_width_list = []
    new_height_list = []
    for idx in range(len(id_class_list)):
        new_x_center = float(x_center_list[idx]) * width_img
        new_x_center_list.append(str(new_x_center))
        
        new_y_center = float(y_center_list[idx]) * height_img
        new_y_center_list.append(str(new_y_center))

        new_width = float(width_list[idx]) * width_img
        new_width_list.append(str(new_width))

        new_height = float(height_list[idx]) * height_img
        new_height_list.append(str(new_height))

    return new_x_center_list, new_y_center_list, new_width_list, new_height_list

def write_real_coor_file(file_path, out_file_path, width_img, height_img, *args, **kwargs):
    ##
    # Steps:
    # 1. Extract contents from the input.txt file.
    # 2. Compute the real coordinates (center X, center Y, width, height).
    # 3. Write resulting information to a new file.
    id_class_list, x_center_list, y_center_list, width_list, height_list = extract_content(file_path)
    new_x_center_list, new_y_center_list, new_width_list, new_height_list = compute_real_coord(id_class_list, x_center_list, y_center_list, width_list, height_list, width_img, height_img)

    with open(out_file_path, 'w') as f:
        for i in range(len(id_class_list)):
            line = id_class_list[i] + ' ' + new_x_center_list[i] + ' ' + new_y_center_list[i] + ' ' + new_width_list[i] + ' ' + new_height_list[i] + '\n'
            f.write(line) 

    return None

def compute_tlbr_coord(id_class_list, new_x_center_list, new_y_center_list, new_width_list, new_height_list, width_img, height_img, *args, **kwargs):
    x_tl_list = []
    y_tl_list = []
    x_br_list = []
    y_br_list = []
    for idx in range(len(id_class_list)):
        x_tl = (1 / width_img) * (float(new_x_center_list[idx]) - float(new_width_list[idx]) / 2)
        x_tl_list.append(str(x_tl))

        y_tl = (1 / height_img) * (float(new_y_center_list[idx]) - float(new_height_list[idx]) / 2)
        y_tl_list.append(str(y_tl))

        x_br = (1 / width_img) * (float(new_x_center_list[idx]) + float(new_width_list[idx]) / 2)
        x_br_list.append(str(x_br))

        y_br = (1 / height_img) * (float(new_x_center_list[idx]) + float(new_height_list[idx]) / 2)
        y_br_list.append(str(y_br))
        
    return x_tl_list, y_tl_list, x_br_list, y_br_list

def write_tlbr_coor_file(file_path, out_file_path, width_img, height_img, *args, **kwargs):
    id_class_list, x_center_list, y_center_list, width_list, height_list = extract_content(file_path)
    new_x_center_list, new_y_center_list, new_width_list, new_height_list = compute_real_coord(id_class_list, x_center_list, y_center_list, width_list, height_list, width_img, height_img)
    x_tl_list, y_tl_list, x_br_list, y_br_list = compute_tlbr_coord(id_class_list, new_x_center_list, new_y_center_list, new_width_list, new_height_list, height_img, width_img)
    
    with open(out_file_path, 'w') as f:
        for i in range(len(id_class_list)):
            line = id_class_list[i] + ' ' + x_tl_list[i] + ' ' + y_tl_list[i] + ' ' + x_br_list[i] + ' ' + y_br_list[i] + '\n'
            f.write(line) 
    
    return None
````
